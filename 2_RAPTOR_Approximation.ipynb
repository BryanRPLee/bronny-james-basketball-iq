# Pseudo-RAPTOR appromixation

## Data Sets

```python
import pandas as pd
import json

# Traditional Box-Scores
with open('./data/traditional.json') as f:
    traditional_json = json.load(f)

headers = traditional_json['resultSets'][0]['headers']
rows = traditional_json['resultSets'][0]['rowSet']

traditional_data = pd.DataFrame(rows, columns=headers)
traditional_data = traditional_data.drop(columns=['PLAYER_ID', 'TEAM_ID', 'NICKNAME'])
traditional_data = traditional_data.loc[:, ~traditional_data.columns.str.contains('RANK|FANTASY', case=False)]

traditional_data.head()
```

```python
# Advanced Stats
with open('./data/advanced.json') as f:
    advanced_json = json.load(f)

headers = advanced_json['resultSets'][0]['headers']
rows = advanced_json['resultSets'][0]['rowSet']

advanced_data = pd.DataFrame(rows, columns=headers)
advanced_data = advanced_data.drop(columns=['PLAYER_ID', 'TEAM_ID', 'NICKNAME'])

advanced_data.head()
```

```python
with open('./data/rapm.json') as f:
    rapm_json = json.load(f)

rapm_data = pd.DataFrame(rapm_json)

rapm_data.head()
```

```python
with open('./data/defense.json') as f:
    defense_json = json.load(f)

headers = defense_json['resultSets'][0]['headers']
rows = defense_json['resultSets'][0]['rowSet']

defense_data = pd.DataFrame(rows, columns=headers)
defense_data = defense_data.drop(columns=['PLAYER_ID', 'TEAM_ID', 'NICKNAME'])

defense_data.head()
```

```python
from unidecode import unidecode

def normalise_name(name):
    return unidecode(name).strip().lower()

traditional_data['PLAYER_NAME'] = traditional_data['PLAYER_NAME'].apply(normalise_name)
advanced_data['PLAYER_NAME'] = advanced_data['PLAYER_NAME'].apply(normalise_name)
defense_data['PLAYER_NAME'] = defense_data['PLAYER_NAME'].apply(normalise_name)
rapm_data['player_name'] = rapm_data['player_name'].apply(normalise_name)

print('Traditional:', 'nikola jokic' in traditional_data['PLAYER_NAME'].values)
print('Advanced:', 'nikola jokic' in advanced_data['PLAYER_NAME'].values)
print('Defense:', 'nikola jokic' in defense_data['PLAYER_NAME'].values)
print('RAPM:', 'nikola jokic' in rapm_data['player_name'].values)
```

```python
merged_data = pd.merge(traditional_data, advanced_data, on='PLAYER_NAME', how='inner')
merged_data = pd.merge(merged_data, defense_data, on='PLAYER_NAME', how='inner')
merged_data = pd.merge(merged_data, rapm_data, left_on='PLAYER_NAME', right_on='player_name', how='inner')
merged_data = merged_data.drop(columns=['player_name'])

merged_data.head()
```

## Model

```python
merged_data['RAPTOR_RATING'] = merged_data['rapm_darko']

merged_data[['PLAYER_NAME', 'RAPTOR_RATING']].sort_values(by='RAPTOR_RATING', ascending=False).head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import numpy as np

# Offense
merged_data['PTS_PER_MIN'] = merged_data['PTS'] / merged_data['MIN']
merged_data['AST_PER_MIN'] = merged_data['AST'] / merged_data['MIN']
merged_data['REB_PER_MIN'] = merged_data['REB'] / merged_data['MIN']

# Efficiency
merged_data['PTS_TS'] = merged_data['PTS'] * merged_data['TS_PCT']
merged_data['AST_TO_USG'] = merged_data['AST_TO'] * merged_data['USG_PCT']

# Defense
merged_data['DEF_PLAYMAKING'] = merged_data['PCT_STL'] + merged_data['PCT_BLK']

# Interaction
merged_data['PTS_AST'] = merged_data['PTS'] * merged_data['AST']
merged_data['REB_AST'] = merged_data['REB'] * merged_data['AST']

features = merged_data[[
    'PTS', 'AST', 'REB', 'TS_PCT', 'AST_TO', 'USG_PCT', 'FG3_PCT', 'FT_PCT',
    'PCT_STL', 'PCT_BLK', 'PCT_DREB', 'rapm_timedecay', 'orapm_timedecay', 'drapm_timedecay',
    'PTS_PER_MIN', 'AST_PER_MIN', 'REB_PER_MIN', 'PTS_TS', 'AST_TO_USG', 'DEF_PLAYMAKING', 'PTS_AST', 'REB_AST'
]]
target = merged_data['RAPTOR_RATING']

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(RandomForestRegressor(random_state=42), param_grid, cv=3, scoring='r2')
grid_search.fit(X_train_scaled, y_train)

model = grid_search.best_estimator_

print("Best Parameters:", grid_search.best_params_)
print("Model R^2 Score:", model.score(X_test_scaled, y_test))
```

```python
predicted_ratings = model.predict(X_test)

X_test['PREDICTED_RAPTOR_RATING'] = predicted_ratings
X_test['ACTUAL_RAPTOR_RATING'] = y_test.values

X_test.head()
```
